# Análisis de Datos Meteorológicos

Este proyecto consiste en analizar un conjunto de datos meteorológicos utilizando Python y NumPy. 
El objetivo es aplicar las técnicas de manipulación de arrays, tratamiento de datos faltantes, 
y análisis estadístico básico en un conjunto de datos del mundo real.


```python
import pandas as pd
import numpy as np
```

### 1. Crear un DataFrame a partir de los datos del archivo datos_meteorologicos.csv.

```python
df= pd.read_csv("datos_meteorologicos.csv")
print(df.head())
```

### 2. Realizar observaciones iniciales de los datos con Pandas.

```python
print(df.info())

print ("\nPodemos observar la existencia de datos vacios dentro de nuestras columnas \n" 
"así como la necesidad de convertir nuestra fecha al tipo de dato correcto")

print(f"\nCantidad de datos Nulos= {df["Temperatura"].isnull().sum()}")
```

```python
#Cambiar formato Fecha
df["Fecha"]=pd.to_datetime(df["Fecha"], format="%d/%m/%Y")
```

### 3. Convertir las columnas del DataFrame en arrays de NumPy.

```python
array=df.to_numpy()
```

### 4. Identificar los datos faltantes en los arrays, y reemplazarlos por el promedio de los valores del respectivo array.

```python
array[:,1:]= np.where(np.isnan(array[:,1:].astype(float)),np.nanmean(array[:,1:].astype(float)),array[:,1:].astype(float))
```

### 5. Realizar análisis estadísticos básicos.

##### Temperatura promedio

```python
TP=np.mean(array[:,1])
print(f"La temperatura promedio es= {TP}")
```

##### Total de precipitaciones

```python
TPR=np.sum(array[:,2])
print(f"El total de precipitaciones es= {TPR}")
```

##### Máxima humedad registrada

```python
MH=np.max(array[:,3])
print(f"El maximo de humedad registrada es= {MH}")
```

##### Fecha más calurosa

```python
MC=np.argmax(array[:,1])
FMC= array[MC,0]
print(f"La fecha más calurosa es= {FMC}")
```

##### Fecha más fría

```python
MF=np.argmin(array[:,1])
FMF= array[MF,0]
print(f"La fecha más calurosa es= {FMF}")
```

```python
#Crear dataframe
Resultados=pd.DataFrame({"Metrica":["Temperatura promedio","Total de precipitaciones","Máxima humedad registrada","Fecha más calurosa","Fecha más fría"],
                         "Resultado":[TP,TPR,MH,FMC,FMF]})

#Exportar
Resultados.to_csv("D:\Curso Udemy machiene learnng\Día+8\Resultados.csv",index=False)
```
