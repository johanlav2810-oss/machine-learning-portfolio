```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
```

 BASES DE DATOS

```python
df1= pd.read_csv("MODELO1.CSV")
df2= pd.read_csv("MODELO2.CSV")
df3= pd.read_csv("MODELO3.CSV")
```

VARIABLES 

```python
x1="dm2"
x2="ganchera"
x3="color"
x4="material"
y="und" # hace refencia al tiempo de operación
```

GRAFICAS MODELO 1

```python
g1= df1[x1]
g2= df1[x2]
g3= df1[x3]
g4= df1[x4]
dy= df1[y]


fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].scatter(g1,dy)
axes[0, 0].set_title("Tiempo vs dm2 ")


axes[0, 1].scatter(g2,dy)
axes[0, 1].set_title("Tiempo vs und/ganchera")

axes[1, 0].scatter(g3,dy)
axes[1, 0].set_title("Tiempo vs color ")

axes[1, 1].scatter(g4,dy)
axes[1, 1].set_title("Tiempo vs material ")

plt.tight_layout()
plt.show()
```
