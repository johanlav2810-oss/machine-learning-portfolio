# Analizador de Ventas

## Objetivo
Crear un programa en Python que analice un conjunto de **datos de ventas de una tienda**. El programa debe realizar varias operaciones de Data Science para proporcionar información valiosa sobre las ventas de la tienda.

### Descripción de los datos 

| Variable | Descripción         |
| -------- | ------------------- |
| Fecha    | Día de la venta     |
| Producto | Familia de productos al que pertenece |
| Cantidad | Unidades vendidas   |
| Ventas   | Valor vendido       |


```python
import pandas as pd
```

Lectura de base de datos 1

```python
df1=pd.read_csv("Datos_Ventas_Tienda.csv")
print(df1.head())
df2=pd.read_csv("Datos_Ventas_Tienda2.csv")
print(df2.head())
```

Fusión de Datos

```python
DF= pd.concat([df1,df2], ignore_index=True)
print(DF)
```

Tratamiento de Datos

```python
#Ver Información General
DF.info()
```

```python
#Valores nulos en el DF
Nulos= DF.isnull().sum()
print(Nulos)

#Valores Duplicados
Duplicados= DF.duplicated().sum()
print(Duplicados)
```

```python
#Eliminar Duplicados DF
DF= DF.drop_duplicates()
print(DF)
```

```python
#Ver Coherencia En venta Total
DF_filtrado= DF[DF["Total Venta"]!= DF["Cantidad"]*DF["Precio Unitario"]]
print(DF_filtrado)
```

```python
#Buscar los Días con Mayores Ventas
DF_Mayor= DF.sort_values("Total Venta",ascending=False)
print(DF_Mayor.head())

#Buscar los Días con Menores Ventas
DF_Menor= DF.sort_values("Total Venta")
print(DF_Menor.head())
```

```python
#Verificar Tipo de Dato Fecha
type(DF["Fecha"][0])
```

```python
#Modificar Tipo de Dato Fecha
DF["Fecha"]=pd.to_datetime(DF["Fecha"], format='%m/%d/%Y')
DF
```

```python
#Añadir Columna Mes
DF["Mes"]= pd.to_datetime(DF["Fecha"]).dt.month
print(DF)
```

Análisis de Ventas

```python
#Producto Más Vendido
DF_PMV= DF.groupby("Producto")["Cantidad"].sum()
DF_PMVO= DF_PMV.sort_values(ascending=False)
print(DF_PMVO)

#Producto Con Mayor Ganancia en Ventas
DF_PMG= DF.groupby("Producto")["Total Venta"].sum()
DF_PMGO= DF_PMG.sort_values(ascending=False)
print(DF_PMGO)

print("\nLa sección de alimentos es la que registra la mayor cantidad de ventas en términos de unidades.\n"
"Sin embargo, al analizar la contribución a las ventas totales,\n"
"los productos eléctricos son los que generan mayores ingresos. De hecho, \n"
"aunque los alimentos representan el mayor volumen de ventas,\n"
"son la categoría que menos dinero aporta al total de ingresos.\n")
```

```python
#Mes con Más Ventas
DF_MMV= DF.groupby("Mes")["Cantidad"].sum()
DF_MMVO= DF_MMV.sort_values(ascending=False)
print(DF_MMVO)

#Mes con Mayor Ganancia
DF_MMG= DF.groupby("Mes")["Total Venta"].sum()
DF_MMGO= DF_MMG.sort_values(ascending=False)
print(DF_MMGO)

print("En el análisis por meses, se observa que los tres meses con mejor desempeño fueron agosto\n" 
"enero y mayo. Estos meses se mantienen entre los de mayor volumen de ventas y, al mismo tiempo,\n" \
"entre los que generan mayores ingresos, lo que evidencia una relación positiva entre la cantidad vendida y\n" \
"la rentabilidad obtenida durante dichos periodos.")
```
